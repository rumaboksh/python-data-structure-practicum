```python
### Q.1 Discuss string slicing and provide examples.

ans)   String slicing is a powerful feature in Python that allows you to extract a subset of characters from a string.## It's done using the following syntax:

string[start:stop:step]

Here:

- start is the initial index (inclusive)
- stop is the final index (exclusive)
- step is the increment between indices (default is 1)

Examples:

1. my_string = "Hello, World!"
    - my_string[0:5] returns "Hello" (from index 0 to 5)
    - my_string[7:] returns "World!" (from index 7 to the end)
    - my_string[:5] returns "Hello" (from the beginning to index 5)
2. my_string = "abcdefg"
    - my_string[1:4] returns "bcd" (from index 1 to 4)
    - my_string[::2] returns "aceg" (every other character, starting from the beginning)
    - my_string[2:6:2] returns "ce" (from index 2 to 6, stepping by 2)

Some important notes:

- If start is omitted, it defaults to 0.
- If stop is omitted, it defaults to the end of the string.
- If step is omitted, it defaults to 1.
- You can use negative indices to count from the end of the string (e.g., -1 refers to the last character).
- You can also use slice notation to extract characters from other sequences, like lists and tuples.

String slicing is a flexible and efficient way to manipulate strings in Python.
```

```python
### Q.2 Explain the Key features of lists in python.

   ans)     In Python, lists are a fundamental data structure that offers the following key features:

1. Ordered collection: Lists store elements in a specific order, and this order is preserved.

2. Mutable: Lists are mutable, meaning you can modify them after creation (e.g., add, remove, or change elements).

3. Dynamic size: Lists can grow or shrink dynamically as elements are added or removed.

4. Heterogeneous: Lists can contain elements of different data types, including strings, integers, floats, and other lists.

5. Indexing: Lists support indexing, which allows accessing specific elements using their index (position in the list).

6. Slicing: Lists support slicing, which enables extracting subsets of elements using a range of indices.

7. Extendability: Lists can be extended using the extend() method or the + operator.

8. Insertion: Elements can be inserted at specific positions using the insert() method.

9. Removal: Elements can be removed using the remove() method or the del statement.

10. Iteration: Lists support iteration, allowing you to loop through their elements using a for loop.

11. Built-in methods: Lists have various built-in methods for common operations like sorting, reversing, and finding the length.

These features make lists a versatile and powerful data structure in Python, suitable for a wide range of applications, from simple collections to complex data processing tasks.
```

```python
### Q.3 Describe how to access, modify,and delete elements in a list.

      ans)   ## Here's how to access, modify, and delete elements in a list:

Accessing Elements:

- Indexing: Use square brackets [] with the index (position) of the element.
- Example: my_list = [1, 2, 3, 4, 5]; print(my_list[0]) # Output: 1

Modifying Elements:

- Assign a new value to an existing index.
- Example: my_list = [1, 2, 3, 4, 5]; my_list[0] = 10; print(my_list) # Output: [10, 2, 3, 4, 5]

Deleting Elements:

- Use the del statement with the index.
- Example: my_list = [1, 2, 3, 4, 5]; del my_list[0]; print(my_list) # Output: [2, 3, 4, 5]
- Use the remove() method to delete the first occurrence of a value.
- Example: my_list = [1, 2, 3, 4, 5]; my_list.remove(3); print(my_list) # Output: [1, 2, 4, 5]

Additional examples:

- Accessing a range of elements: my_list[1:3] # Output: [2, 3]
- Modifying a range of elements: my_list[1:3] = [20, 30]; print(my_list) # Output: [1, 20, 30, 4, 5]
- Deleting a range of elements: del my_list[1:3]; print(my_list) # Output: [1, 4, 5]

Note: List indices start at 0, and negative indices count from the end of the list (e.g., -1 refers to the last element).
```

```python
### Q.4 Compare and contrast Tuples and lists with examples.

ans)    ## Tuples and lists are both data structures in Python, but they have some key differences:

Similarities:

- Both can store multiple values in a single variable.
- Both support indexing and slicing.
- Both can be used to store a collection of data.

Differences:

- Immutability: Tuples are immutable, meaning their contents cannot be changed after creation. Lists are mutable, meaning their contents can be modified.
- Syntax: Tuples use parentheses () while lists use square brackets [].
- Usage: Tuples are often used for data that shouldn't be changed, like coordinates or dates, while lists are used for dynamic collections.

Examples:

- Tuple:
    - my_tuple = (1, 2, 3)
    - my_tuple[0] = 10 # Error! Tuples are immutable
- List:
    - my_list = [1, 2, 3]
    - my_list[0] = 10 # Okay! Lists are mutable

Other key differences:

- Tuples are faster and more memory-efficient than lists.
- Tuples can be used as dictionary keys, while lists cannot.
- Tuples have a fixed size, while lists can grow or shrink dynamically.

```

```python
### Q.5 Describe the key features of sets and provide examples of their use.

 ans) Sets are a fundamental data structure in Python, offering the following key features:

1. Unordered collection: Sets store elements in no particular order.
2. No duplicates: Sets automatically remove duplicate elements.
3. Mutable: Sets can be modified after creation (elements can be added or removed).
4. Fast membership testing: Checking if an element is in a set is very efficient.

Examples of set usage:

1. Removing duplicates from a list:
    - my_list = [1, 2, 2, 3, 4, 4, 5]; my_set = set(my_list); print(my_set) # Output: {1, 2, 3, 4, 5}
2. Checking membership:
    - my_set = {1, 2, 3}; print(2 in my_set) # Output: True
3. Set operations:
    - Union: set1 = {1, 2, 3}; set2 = {3, 4, 5}; print(set1 | set2) # Output: {1, 2, 3, 4, 5}
    - Intersection: set1 = {1, 2, 3}; set2 = {3, 4, 5}; print(set1 & set2) # Output: {3}
    - Difference: set1 = {1, 2, 3}; set2 = {3, 4, 5}; print(set1 - set2) # Output: {1, 2}
4. Set comprehension:
    - my_set = {x**2 for x in range(5)]; print(my_set) # Output: {0, 1, 4, 9, 16}

Sets are useful when you need to:

- Remove duplicates from a collection
- Check membership quickly
- Perform set operations (union, intersection, difference)
- Create a collection of unique elements
```

```python
### Q.6 Discuss the use cases of tuples and sets in python programming.

ans)  Tuples and sets are both fundamental data structures in Python, and they have different use cases:

Tuples:

1. Immutable data: Use tuples when you need to store data that shouldn't be changed, like coordinates, dates, or fixed settings.
2. Small collections: Tuples are suitable for small collections of data, like a pair of values or a small set of constants.
3. Function arguments: Tuples are often used as function arguments when you need to pass a small, fixed number of values.
4. Data storage: Tuples can be used to store data in a compact, immutable format, like a database record or a configuration file.
5. Performance: Tuples are faster and more memory-efficient than lists for small collections.

Sets:

1. Unique values: Use sets when you need to store a collection of unique values, like a list of IDs or a set of keywords.
2. Fast membership testing: Sets are ideal for quick membership testing, like checking if a value is in a collection.
3. Set operations: Sets support union, intersection, and difference operations, making them perfect for complex data analysis.
4. Data deduplication: Sets can be used to remove duplicates from a large dataset.
5. Caching: Sets can be used as a cache to store unique values and quickly check membership.

In summary:

- Tuples are ideal for small, immutable collections of data.
- Sets are perfect for storing unique values and performing set operations.

When to choose tuples over sets:

- When data is immutable and small.
- When performance is critical.

When to choose sets over tuples:

- When data is large and needs to be deduplicated.
- When fast membership testing is essential.
```

```python
### Q.7 Describe how to add, modify and delete items in a dictionary with examples.

ans) Dictionaries in Python are mutable data structures that allow you to add, modify, and delete items. Here are some examples:

Adding items:

You can add a new item to a dictionary by assigning a value to a new key. If the key doesn't exist, it will be created.

Example:

my_dict = {'a': 1, 'b': 2}
my_dict['c'] = 3
print(my_dict)  # Output: {'a': 1, 'b': 2, 'c': 3}


Modifying items:

You can modify an existing item by assigning a new value to an existing key.

Example:

my_dict = {'a': 1, 'b': 2}
my_dict['a'] = 10
print(my_dict)  # Output: {'a': 10, 'b': 2}


Deleting items:

You can delete an item using the del statement or the pop() method.

Example using del:

my_dict = {'a': 1, 'b': 2}
del my_dict['a']
print(my_dict)  # Output: {'b': 2}


Example using pop():

my_dict = {'a': 1, 'b': 2}
my_dict.pop('a')
print(my_dict)  # Output: {'b': 2}


Note that if you try to delete a key that doesn't exist, you'll get a KeyError. You can use the pop() method with a default value to avoid this:

my_dict = {'a': 1, 'b': 2}
my_dict.pop('c', None)  # Returns None instead of raising a KeyError
```

```python
### Q.8 Discuss the importance of dictionary keys being immutable and provide examples.

ans) n Python, dictionary keys must be immutable, meaning they cannot be changed after creation. This is important for several reasons:

1. Hashing: Dictionaries use hashing to store and retrieve key-value pairs efficiently. Immutable keys ensure that the hash value remains constant, allowing for fast lookups.
2. Uniqueness: Immutable keys guarantee uniqueness, preventing duplicate keys in a dictionary.
3. Consistency: Immutable keys ensure that the dictionary remains consistent, even when modifying or deleting key-value pairs.

Examples of immutable key types:

- Integers (e.g., 1, 2, 3)
- Floats (e.g., 3.14, -0.5)
- Strings (e.g., "hello", "goodbye")
- Tuples (e.g., (1, 2, 3), ("a", "b", "c"))

Examples of mutable key types (which cannot be used as dictionary keys):

- Lists (e.g., [1, 2, 3], ["a", "b", "c"])
- Dictionaries (e.g., {"a": 1, "b": 2})
- Sets (e.g., {1, 2, 3}, {"a", "b", "c"})

If you try to use a mutable object as a dictionary key, you'll get a TypeError. For example:

my_dict = {}
my_list = [1, 2, 3]
my_dict[my_list] = "value"  # Raises TypeError: unhashable type: 'list'

To use a mutable object as a key, you can convert it to an immutable type, such as a tuple:

my_dict = {}
my_list = [1, 2, 3]
my_dict[tuple(my_list)] = "value"  # Works!

In summary, the immutability of dictionary keys is crucial for maintaining the integrity and efficiency of dictionaries in Python.
```
